```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# 设置随机种子以确保结果可重现
np.random.seed(42)

# 定义可能的机型
models = ['iPhone 12', 'iPhone 12 Pro', 'iPhone 12 Pro Max', 'iPhone 13', 
          'iPhone 13 Pro', 'iPhone 13 Pro Max', 'iPhone 14', 'iPhone 14 Pro']

# 定义可能的存储容量
capacities = [64, 128, 256, 512, 1024]  # 单位：GB

# 定义可能的成色及其中东和欧洲的分布概率
conditions = ['准新', '95新', '9成新', '85新', '8成新']
# 准新机型在中东的分布概率更高
middle_east_condition_probs = [0.4, 0.25, 0.15, 0.1, 0.1]
europe_condition_probs = [0.2, 0.25, 0.25, 0.15, 0.15]

# 定义交易国家及其所属区域
countries = {
    '中东': ['沙特阿拉伯', '阿联酋', '卡塔尔', '科威特', '以色列'],
    '欧洲': ['英国', '德国', '法国', '意大利', '西班牙', '荷兰', '波兰']
}

# 基本价格表（准新成色的基础价格）
base_prices = {
    'iPhone 12': 400,
    'iPhone 12 Pro': 500,
    'iPhone 12 Pro Max': 550,
    'iPhone 13': 500,
    'iPhone 13 Pro': 650,
    'iPhone 13 Pro Max': 700,
    'iPhone 14': 600,
    'iPhone 14 Pro': 800
}

# 成色对价格的影响系数
condition_price_factors = {
    '准新': 1.0,
    '95新': 0.9,
    '9成新': 0.8,
    '85新': 0.7,
    '8成新': 0.6
}

# 存储容量对价格的影响系数
capacity_price_factors = {
    64: 0.8,
    128: 0.9,
    256: 1.0,
    512: 1.3,
    1024: 1.8
}

# 生成5000条记录
data = []
for _ in range(5000):
    # 随机选择区域（中东或欧洲），控制中东占比60%
    region = '中东' if np.random.random() < 0.6 else '欧洲'
    
    # 随机选择国家
    country = np.random.choice(countries[region])
    
    # 随机选择机型
    model = np.random.choice(models)
    
    # 根据区域调整容量分布（欧洲中东集中为256GB及以下机型）
    if region == '中东':
        # 中东地区256GB及以下占80%
        if np.random.random() < 0.8:
            capacity = np.random.choice(capacities[:3])  # 64, 128, 256
        else:
            capacity = np.random.choice(capacities[3:])  # 512, 1024
    else:
        # 欧洲地区256GB及以下占70%
        if np.random.random() < 0.7:
            capacity = np.random.choice(capacities[:3])  # 64, 128, 256
        else:
            capacity = np.random.choice(capacities[3:])  # 512, 1024
    
    # 根据区域调整成色分布
    if region == '中东':
        condition = np.random.choice(conditions, p=middle_east_condition_probs)
    else:
        condition = np.random.choice(conditions, p=europe_condition_probs)
    
    # 计算基础价格
    base_price = base_prices[model] * condition_price_factors[condition] * capacity_price_factors[capacity]
    
    # 根据区域调整价格（中东价格高出20%）
    if region == '中东':
        price = base_price * 1.2
    else:
        price = base_price
    
    # 添加随机波动（±10%）
    price = price * (1 + np.random.uniform(-0.1, 0.1))
    
    # 计算销售周期（中东销售周期较短）
    if region == '中东':
        # 中东销售周期在7-15天之间
        sales_cycle = np.random.randint(7, 16)
    else:
        # 欧洲销售周期在12-25天之间
        sales_cycle = np.random.randint(12, 26)
    
    # 生成交易日期（过去一年内的随机日期）
    days_ago = np.random.randint(0, 366)
    transaction_date = (datetime.now() - timedelta(days=days_ago)).strftime('%Y-%m-%d')
    
    # 生成批发数量（以20台为一箱）
    boxes = np.random.randint(1, 51)  # 1-50箱
    quantity = boxes * 20
    
    # 计算总金额
    total_amount = price * quantity
    
    # 添加记录到数据列表
    data.append({
        '交易日期': transaction_date,
        '交易国家': country,
        '所属区域': region,
        '机型': model,
        '存储容量(GB)': capacity,
        '成色': condition,
        '单价(美元)': round(price, 2),
        '数量(台)': quantity,
        '销售周期(天)': sales_cycle,
        '总金额(美元)': round(total_amount, 2)
    })

# 创建DataFrame
df = pd.DataFrame(data)

# 保存数据到CSV文件
df.to_csv('二手苹果手机跨国批发数据.csv', index=False, encoding='utf-8-sig')

print(f"成功生成{len(df)}条模拟数据并保存为CSV文件。")    
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 读取数据
try:
    df = pd.read_csv('二手苹果手机跨国批发数据.csv')
except FileNotFoundError:
    print("错误：找不到数据文件，请确保第一步生成的数据文件存在。")
    exit()

# 设置中文字体
plt.rcParams["font.family"] = ["SimHei", "WenQuanYi Micro Hei", "Heiti TC"]
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 1. 缺失值检查
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
missing_data = pd.DataFrame({
    '列名': df.columns,
    '缺失值数量': missing_values,
    '缺失值百分比': missing_percentage.round(2).astype(str) + '%'
})

# 2. 异常值检测
numeric_columns = ['单价(美元)', '数量(台)', '销售周期(天)', '总金额(美元)']
outliers = {}

for col in numeric_columns:
    # 使用IQR方法检测异常值
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # 计算异常值数量
    outliers_count = len(df[(df[col] < lower_bound) | (df[col] > upper_bound)])
    outliers_percentage = (outliers_count / len(df)) * 100
    
    outliers[col] = {
        '下界': round(lower_bound, 2),
        '上界': round(upper_bound, 2),
        '异常值数量': outliers_count,
        '异常值百分比': round(outliers_percentage, 2)
    }
    
    # 为后续分析，标记这些异常值
    df[f'{col}_outlier'] = (df[col] < lower_bound) | (df[col] > upper_bound)

# 3. 数据一致性检查
# 检查单价乘以数量是否等于总金额
df['金额计算验证'] = np.abs(df['单价(美元)'] * df['数量(台)'] - df['总金额(美元)']) > 0.01
calculation_errors = df['金额计算验证'].sum()
calculation_errors_percentage = (calculation_errors / len(df)) * 100

# 检查存储容量是否符合预期值
valid_capacities = [64, 128, 256, 512, 1024]
invalid_capacities = df[~df['存储容量(GB)'].isin(valid_capacities)]

# 检查成色是否符合预期值
valid_conditions = ['准新', '95新', '9成新', '85新', '8成新']
invalid_conditions = df[~df['成色'].isin(valid_conditions)]

# 检查区域与国家是否匹配
region_country_mapping = {
    '中东': ['沙特阿拉伯', '阿联酋', '卡塔尔', '科威特', '以色列'],
    '欧洲': ['英国', '德国', '法国', '意大利', '西班牙', '荷兰', '波兰']
}

country_region_dict = {}
for region, countries in region_country_mapping.items():
    for country in countries:
        country_region_dict[country] = region

invalid_region_country = 0
for index, row in df.iterrows():
    if row['交易国家'] in country_region_dict and country_region_dict[row['交易国家']] != row['所属区域']:
        invalid_region_country += 1

# 4. 数据分布概述
numeric_summary = df[numeric_columns].describe().round(2)

# 5. 可视化分析
plt.figure(figsize=(15, 10))

# 绘制箱线图检测异常值
plt.subplot(2, 2, 1)
sns.boxplot(data=df[numeric_columns])
plt.title('数值型字段箱线图')

# 绘制直方图查看分布
plt.subplot(2, 2, 2)
df[numeric_columns].hist(ax=plt.gca())
plt.tight_layout()

# 绘制散点图查看单价与总金额关系
plt.subplot(2, 2, 3)
sns.scatterplot(x='单价(美元)', y='总金额(美元)', hue='所属区域', data=df)
plt.title('单价与总金额散点图')

# 绘制热力图查看相关性
plt.subplot(2, 2, 4)
correlation = df[numeric_columns].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('数值型字段相关性热力图')

plt.tight_layout()
plt.savefig('数据质量评估可视化.png')

# 6. 输出质量评估结果
print("数据质量评估报告：")
print("\n1. 缺失值检查：")
print(missing_data)

print("\n2. 异常值检测：")
for col, outlier_info in outliers.items():
    print(f"\n{col}:")
    print(f"  异常值判定范围: ({outlier_info['下界']}, {outlier_info['上界']})")
    print(f"  异常值数量: {outlier_info['异常值数量']} ({outlier_info['异常值百分比']}%)")

print("\n3. 数据一致性检查：")
print(f"  金额计算错误数量: {calculation_errors} ({calculation_errors_percentage}%)")
print(f"  无效存储容量数量: {len(invalid_capacities)}")
print(f"  无效成色数量: {len(invalid_conditions)}")
print(f"  区域与国家不匹配数量: {invalid_region_country}")

print("\n4. 数据分布概述：")
print(numeric_summary)

# 保存评估结果到CSV
missing_data.to_csv('缺失值检查结果.csv', index=False)
pd.DataFrame(outliers).T.to_csv('异常值检测结果.csv')
numeric_summary.to_csv('数据分布概述.csv')

print("\n数据质量评估完成，结果已保存到文件。")    
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from statsmodels.tsa.arima.model import ARIMA
import warnings
warnings.filterwarnings('ignore')

# 设置中文字体
plt.rcParams["font.family"] = ["SimHei", "WenQuanYi Micro Hei", "Heiti TC"]
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 读取数据
try:
    df = pd.read_csv('二手苹果手机跨国批发数据.csv')
except FileNotFoundError:
    print("错误：找不到数据文件，请确保数据文件存在。")
    exit()

# 数据预处理
df['交易日期'] = pd.to_datetime(df['交易日期'])
df['月份'] = df['交易日期'].dt.to_period('M')

# 计算每月市场规模（总销售额）
monthly_market_size = df.groupby('月份')['总金额(美元)'].sum().reset_index()
monthly_market_size['月份'] = monthly_market_size['月份'].astype(str)

# 计算每月利润（假设利润率为15%）
monthly_market_size['利润(美元)'] = monthly_market_size['总金额(美元)'] * 0.15

# 预测模型 - 市场规模
# 准备时间序列数据
ts_data = monthly_market_size.set_index('月份')['总金额(美元)']

# 使用ARIMA模型进行预测
try:
    # 拟合ARIMA模型
    model = ARIMA(ts_data, order=(1, 1, 1))
    model_fit = model.fit()
    
    # 预测未来6个月
    forecast_steps = 6
    forecast = model_fit.forecast(steps=forecast_steps)
    
    # 创建未来日期
    last_date = pd.to_datetime(ts_data.index[-1])
    future_dates = [(last_date + pd.DateOffset(months=i+1)).strftime('%Y-%m') for i in range(forecast_steps)]
    
    # 可视化市场规模预测
    plt.figure(figsize=(12, 6))
    plt.plot(ts_data.index, ts_data, label='历史数据')
    plt.plot(future_dates, forecast, 'r--', label='预测数据')
    plt.title('未来6个月市场规模预测')
    plt.xlabel('月份')
    plt.ylabel('市场规模(美元)')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.savefig('市场规模预测.png')
    
    print("\n市场规模预测结果：")
    for date, value in zip(future_dates, forecast):
        print(f"{date}: {value:.2f} 美元")
        
except Exception as e:
    print(f"ARIMA模型预测失败: {e}")
    print("使用线性回归模型进行预测")
    
    # 使用线性回归进行预测
    monthly_market_size['月份_num'] = range(1, len(monthly_market_size) + 1)
    
    X = monthly_market_size[['月份_num']]
    y = monthly_market_size['总金额(美元)']
    
    model = LinearRegression()
    model.fit(X, y)
    
    # 预测未来6个月
    last_month_num = monthly_market_size['月份_num'].iloc[-1]
    future_months = [[last_month_num + i] for i in range(1, forecast_steps + 1)]
    forecast = model.predict(future_months)
    
    # 创建未来日期
    last_date = pd.to_datetime(monthly_market_size['月份'].iloc[-1])
    future_dates = [(last_date + pd.DateOffset(months=i+1)).strftime('%Y-%m') for i in range(forecast_steps)]
    
    # 可视化市场规模预测
    plt.figure(figsize=(12, 6))
    plt.plot(monthly_market_size['月份'], y, label='历史数据')
    plt.plot(future_dates, forecast, 'r--', label='预测数据')
    plt.title('未来6个月市场规模预测（线性回归）')
    plt.xlabel('月份')
    plt.ylabel('市场规模(美元)')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.savefig('市场规模预测_线性回归.png')
    
    print("\n市场规模预测结果（线性回归）：")
    for date, value in zip(future_dates, forecast):
        print(f"{date}: {value:.2f} 美元")

# 利润分析
# 计算各区域、各机型的利润贡献
region_profit = df.groupby('所属区域')['总金额(美元)'].sum().reset_index()
region_profit['利润(美元)'] = region_profit['总金额(美元)'] * 0.15

model_profit = df.groupby('机型')['总金额(美元)'].sum().reset_index()
model_profit['利润(美元)'] = model_profit['总金额(美元)'] * 0.15

# 可视化利润分析
plt.figure(figsize=(15, 10))

# 区域利润分布
plt.subplot(2, 2, 1)
plt.pie(region_profit['利润(美元)'], labels=region_profit['所属区域'], autopct='%1.1f%%', startangle=90)
plt.title('各区域利润分布')

# 机型利润分布
plt.subplot(2, 2, 2)
plt.pie(model_profit['利润(美元)'], labels=model_profit['机型'], autopct='%1.1f%%', startangle=90)
plt.title('各机型利润分布')

# 利润趋势
plt.subplot(2, 2, 3)
monthly_profit = monthly_market_size[['月份', '利润(美元)']]
plt.plot(monthly_profit['月份'], monthly_profit['利润(美元)'])
plt.title('月度利润趋势')
plt.xlabel('月份')
plt.ylabel('利润(美元)')
plt.xticks(rotation=45)

# 单价与利润散点图
plt.subplot(2, 2, 4)
sns.scatterplot(x='单价(美元)', y='总金额(美元)', hue='所属区域', data=df)
plt.title('单价与销售额关系')

plt.tight_layout()
plt.savefig('利润分析.png')

# 保存预测结果
forecast_df = pd.DataFrame({
    '月份': future_dates,
    '市场规模预测(美元)': forecast,
    '利润预测(美元)': forecast * 0.15
})

forecast_df.to_csv('市场与利润预测结果.csv', index=False)

print("\n利润分析结果已保存至图表")
print("市场与利润预测结果已保存至 CSV 文件")    
```
